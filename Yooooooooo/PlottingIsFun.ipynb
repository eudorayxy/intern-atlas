```python
import sys 
sys.path.append('./backend')
from PlotHistogramAk import plot_histogram
```

```python
import pickle

# Load
with open('all_data.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
GeV = 1.0
# x-axis range of the plot
xmin = 0 * GeV
xmax = 120 * GeV

# Histogram bin setup
step_size = 1 * GeV
fig, ax = plot_histogram(data, 'mass',
                         xmin, xmax,
                         step_size,
                         '2-lepton invariant mass $\mathrm{m_{2l}}$ [GeV]',
                         marker='.',
                         label_fontsize=16,
                         title='$Z\\rightarrow\\mu\\mu$',
                         logy=True)
```
